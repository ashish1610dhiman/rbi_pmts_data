#### Agg all months into one

```python
def clean_month(df,mon_key):
    pmt_types = df.iloc[3].dropna().drop_duplicates().values
    tuples = [(pmt_i,data_type)\
         for pmt_i in pmt_types for data_type in ["Vol","Val"]]
    temp = df[5:]
    temp.index = temp["Data for the day"]
    temp.drop(columns="Data for the day",inplace=True)
    temp.columns = pd.MultiIndex.from_tuples(tuples,names=["pmt_type","data_type"])
    try:
        cleaned = temp.iloc[:temp.index.str.contains("Total").get_loc(True)]
    except:
        cleaned = pd.DataFrame()
        print ("***** No exact key found, Skipping *****")
    return cleaned
```

```python
cleaned_all = pd.DataFrame()
for mon_key,df_mon in pmts_data.items():
    cleaned_month = clean_month(df_mon,mon_key)
    print(f"Cleaned data for {mon_key} has shape:{cleaned_month.shape}")
    cleaned_all = cleaned_all.append(cleaned_month)
```

##### Manual for Jul 2020
